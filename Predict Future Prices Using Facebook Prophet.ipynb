# TASK 4: PREPARE THE DATA BEFORE APPLYING FACEBOOK PROPHET TOOL 

```python
avocado_tf
```

```python
avocado_prophet_tf=avocado_tf[['Date','AveragePrice']]
```

```python
avocado_prophet_tf
```

```python
avocado_prophet_tf=avocado_prophet_tf.rename(columns={'Date':'ds','AveragePrice':'y'})
```

```python
avocado_prophet_tf
```

# TASK 5: UNDERSTAND INTUITION BEHIND FACEBOOK PROPHET

# TASK 6: DEVELOP MODEL AND MAKE PREDICTIONS - PART A

```python
m=Prophet()
m.fit(avocado_prophet_tf)
```

```python
# Forcasting into the future
future = m.make_future_dataframe(periods = 365)
forecast = m.predict(future)
```

```python
forecast 
```

```python
figure = m.plot(forecast,xlabel='Data',ylabel='Price')
```

```python
figure2 = m.plot_components(forecast)
```

# TASK 7: DEVELOP MODEL AND MAKE PREDICTIONS (REGION SPECIFIC) - PART B

```python
# dataframes creation for both training and testing datasets 
avocado_tf = pd.read_csv('avocado.csv')
```

```python
# Select specific region
avocado_tf_sample = avocado_tf[avocado_tf['region']=='West']
```

```python
avocado_tf_sample = avocado_tf_sample.sort_values('Date')
```

```python
plt.plot(avocado_tf_sample['Date'],avocado_tf_sample['AveragePrice'])
```

```python
avocado_tf_sample = avocado_tf_sample.rename(columns={'Date':'ds','AveragePrice':'y'})
```

```python
m = Prophet()
m.fit(avocado_tf_sample)
# Forcasting into the future
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
```

```python
figure = m.plot(forecast, xlabel='Date', ylabel='Price')
```

```python
figure3 = m.plot_components(forecast)
```
